```python
%matplotlib
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

wifi_building = pd.DataFrame.from_csv('/Users/mac/Desktop/wifi_trip.csv')     # wifi data
building = gpd.read_file("/Users/mac/Desktop/utm_test/Buildings.shp")         # geodataframe without frequency
Building = pd.DataFrame.from_csv('/Users/mac/Desktop/utm_test/Buildings.csv') # dataframe with frequency

Lurie = building.geometry[14].coords[0]  # fixed bus-stop Lurie
lat_line = 290560                       # line between North Campus and NCRCï¼Œ projection100000
df1 = building[building.Latitude < lat_line]         # buildings in North Campus,geodataframe
df2 = Building[Building.Latitude < lat_line]         # buildings in North Campus,dataframe
```

```python
def dist(n0, n1):
    x1, y1 = n0
    x2, y2 = n1
    distance = np.sqrt((x2-x1)**2 + (y2-y1)**2)
    return distance
def nearest(n0, n1, n2):          # find nearest bus-stop of 2stops
    x0, y0 = n0
    x1, y1 = n1
    x2, y2 = n2
    d1 = dist(n0, n1)
    d2 = dist(n0, n2)
    return min(d1, d2)
def nearest3(n0, n1, n2, n3):     # find nearest bus-stop of 3stops
    x0, y0 = n0
    x1, y1 = n1
    x2, y2 = n2
    x3, y3 = n3
    d1 = dist(n0, n1)
    d2 = dist(n0, n2)
    d3 = dist(n0, n3)
    return min(d1, d2, d3)
def nearest4(n0, n1, n2, n3, n4):     # find nearest bus-stop of 4stops
    x0, y0 = n0
    x1, y1 = n1
    x2, y2 = n2
    x3, y3 = n3
    x4, y4 = n4
    d1 = dist(n0, n1)
    d2 = dist(n0, n2)
    d3 = dist(n0, n3)
    d4 = dist(n0, n4)
    return min(d1, d2, d3, d4)
```

```python
# one bus-stop---------------------------------------------
n1 = Lurie
total_distance = 0
for i in df1.index[(df1.index!=14)]:
    n0 = df1.geometry[i].coords[0]
    total_distance = total_distance + dist(n0, n1)* df2.frequency[i]
average_walktime = total_distance/(sum(df2.frequency)*1.5)
print('average walktime is:'+ str(average_walktime))
```

```python
# two bus-stops--------------------------------------------
total_distance = float('inf')
stop_candidate = 0
n1 = Lurie
for stp in df1.index[(df1.index!=14)&(df1.index!=31)&(df1.index!=2)&(df1.index!=21)&(df1.index!=23)&(df1.index!=24)]:
    dist_total_tmp = 0
    n2 = df1.geometry[stp].coords[0]
    for i in df1.index[(df1.index!=14)&(df1.index!=stp)]:
        n0 = df1.geometry[i].coords[0]
        dist_total_tmp = dist_total_tmp + nearest(n0, n1, n2)* df2.frequency[i]
    if dist_total_tmp < total_distance: 
        total_distance = dist_total_tmp
        stop_candidate = stp
average_walktime = total_distance/(sum(df2.frequency)*1.5)
print('average walktime is:'+ str(average_walktime), "and the candidate stop is building:"+ str(stop_candidate))
```

```python
# three bus-stops--------------------------------------------
total_distance = float('inf')
stop_candidate1 = 0
stop_candidate2 = 0
stp1 = Lurie
for stp2 in df1.index[(df1.index!=14)]:
    dist_total_tmp = float('inf')
    n2 = df1.geometry[stp2].coords[0]
    for stp3 in df1.index[(df1.index!=14)&(df1.index > stp2)]:
        tmp = 0
        n3 = df1.geometry[stp3].coords[0]
        for i in df1.index[(df1.index!=14)&(df1.index!=stp2)&(df1.index!=stp3)]:
            n0 = df1.geometry[i].coords[0]
            tmp = tmp + nearest3(n0, n1, n2, n3)* df2.frequency[i]
        if tmp < dist_total_tmp: 
            dist_total_tmp = tmp
            stop_tmp2 = stp3
    if dist_total_tmp < total_distance:
        stop_candidate1 = stp2
        stop_candidate2 = stop_tmp2
        total_distance = dist_total_tmp
average_walktime = total_distance/(sum(df2.frequency)*1.5)
print('average walktime is:'+ str(average_walktime), 
      "and the candidate stop is building:"+ str(stop_candidate1),'and '+ str(stop_candidate2))
```

```python
# four bus-stops--------------------------------------------
total_distance = float('inf')
stop_candidate1 = 0
stop_candidate2 = 0
stop_candidate3 = 0
stp1 = Lurie
for stp2 in df1.index[(df1.index!=14)]:
    dist_total_tmp = float('inf')
    n2 = df1.geometry[stp2].coords[0]
    for stp3 in df1.index[(df1.index!=14)&(df1.index > stp2)]:
        tmp1 = float('inf')
        n3 = df1.geometry[stp3].coords[0]
        for stp4 in df1.index[(df1.index!=14)&(df1.index > stp3)]:
            tmp2 = 0
            n4 = df1.geometry[stp4].coords[0]
            for i in df1.index[(df1.index!=14)&(df1.index!=stp2)&(df1.index!=stp3)&(df1.index!=stp4)]:
                n0 = df1.geometry[i].coords[0]
                tmp2 = tmp2 + nearest4(n0, n1, n2, n3, n4)* df2.frequency[i]
            if tmp2 < tmp1:
                tmp1 = tmp2
                stop_tmp3 = stp4
        if tmp1 < dist_total_tmp: 
            dist_total_tmp = tmp1
            stop_tmp2 = stp3
    if dist_total_tmp < total_distance:
        stop_candidate1 = stp2
        stop_candidate2 = stop_tmp2
        stop_candidate3 = stop_tmp3
        total_distance = dist_total_tmp
average_walktime = total_distance/(sum(df2.frequency)*1.5)
print('average walktime is:'+ str(average_walktime), 
      "and the candidate stop is building:"+ 
      str(stop_candidate1),'and '+ str(stop_candidate2),'and '+ str(stop_candidate3))
```

```python
# plot----------walking time to Lurie bus-stop
n1 = Lurie
walktime = [0]*df2.frequency[14]
for i in df1.index[(df1.index!=14)]:
    n0 = df1.geometry[i].coords[0]
    t = dist(n0, n1)/1.5
    walktime = walktime+[t]*df2.frequency[i]
hist, bins = np.histogram(walktime, bins=15)
width = 0.6 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center', width=width,alpha=0.7)
plt.xlabel('Walking Time to nearest bus-stop, sec')
plt.ylabel('Frequency')
plt.xlim(0,500)
plt.axvline(253.58, color='b', linestyle='--',linewidth=3.0,label='1 stop:average walking time')
plt.legend(loc='upper right')
plt.show()
```

```python
# plot----------walking time to bus-stops, 2 stops
n1 = Lurie
n2 = df1.geometry[19].coords[0]
walktime = [0]*(df2.frequency[14]+df2.frequency[19])
for i in df1.index[(df1.index!=14)&(df1.index!=19)]:
    n0 = df1.geometry[i].coords[0]
    t = nearest(n0, n1, n2)/1.5
    walktime = walktime+[t]*df2.frequency[i]
hist, bins = np.histogram(walktime, bins=15)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center', width=width,alpha=0.7)
plt.xlabel('Walking Time to nearest bus-stop, sec')
plt.ylabel('Frequency')
plt.xlim(0,500)
plt.axvline(121.33, color='orange', linestyle='--',linewidth=3.0,label='2 stops:average walking time')
plt.legend(loc='upper right')
plt.show()
```

```python
# plot----------walking time to bus-stops, 3 stops
n1 = Lurie
n2 = df1.geometry[33].coords[0]
n3 = df1.geometry[35].coords[0]
walktime = [0]*(df2.frequency[14]+df2.frequency[33]+df2.frequency[35])
for i in df1.index[(df1.index!=14)&(df1.index!=33)&(df1.index!=35)]:
    n0 = df1.geometry[i].coords[0]
    t = nearest3(n0, n1, n2, n3)/1.5
    walktime = walktime+[t]*df2.frequency[i]
hist, bins = np.histogram(walktime, bins=15)
width = 1.0 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center', width=width,alpha=0.7)
plt.xlabel('Walking Time to nearest bus-stop, sec')
plt.ylabel('Frequency')
plt.xlim(0,500)
plt.axvline(74.45, color='green', linestyle='--',linewidth=3.0,label='3 stops:average walking time')
plt.legend(loc='upper right')
plt.show()
```

```python
# plot----------walking time to bus-stops, 4 stops
n1 = Lurie
n2 = df1.geometry[31].coords[0]
n3 = df1.geometry[35].coords[0]
n3 = df1.geometry[60].coords[0]
walktime = [0]*(df2.frequency[14]+df2.frequency[31]+df2.frequency[35]+df2.frequency[60])
for i in df1.index[(df1.index!=14)&(df1.index!=31)&(df1.index!=35)&(df1.index!=60)]:
    n0 = df1.geometry[i].coords[0]
    t = nearest4(n0, n1, n2, n3, n4)/1.5
    walktime = walktime+[t]*df2.frequency[i]
hist, bins = np.histogram(walktime, bins=15)
width = 1.1 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center', width=width,alpha=0.7)
plt.xlabel('Walking Time to Nearest Bus-stop, sec')
plt.ylabel('Frequency')
plt.xlim(0,500)
plt.axvline(58.54, color='red', linestyle='--',linewidth=3.0,label='4 stops:average walking time')
plt.legend(loc='upper right')
plt.show()
```
